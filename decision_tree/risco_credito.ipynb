```python
import pandas as pd
import numpy as np
import pickle 
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
with open(r'C:\Users\natha\Documents\02-Cursos\Udemy - Data Science\bases\risco_credito.pkl', 'rb') as f:
  x_risco_credito, y_risco_credito = pickle.load(f)
```

```python
arvore = DecisionTreeClassifier(criterion='entropy')
arvore.fit(x_risco_credito, y_risco_credito)
```

```python
# Retorna a importancia de cada um dos valores
arvore.feature_importances_
```

```python
# print da arvore em txt e em imagem branca e preta
from sklearn import tree

previsores = ['historia', 'divida', 'garantias', 'renda']
figura, eixos =  plt.subplots(nrows=1, ncols=1, figsize=(30,15))

tree.plot_tree(arvore, feature_names=previsores, class_names=arvore.classes_, filled=True);
```

```python
# Dois testes realizados na parte teorica do curso
# história boa (0), dívida alta (0), garantias nenhuma (1), renda > 35 (2)
# história ruim (2), dívida alta (0), garantias adequada (0), renda < 15 (0)
previsoes = arvore. predict([[0,0,1,2], [2,0,0,0]])
previsoes
```
