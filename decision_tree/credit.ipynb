```python
import pandas as pd
import numpy as np
import pickle 
from sklearn.tree import DecisionTreeClassifier
```

```python
with open(r'C:\Users\natha\Documents\02-Cursos\Udemy - Data Science\bases\credit.pkl', 'rb') as f:
    x_credit_treinamento, x_credit_teste, y_credit_teste, y_credit_treinamento = pickle.load(f)
```

```python
x_credit_treinamento.shape, x_credit_teste.shape
```

```python
y_credit_treinamento.shape, y_credit_teste.shape
```

```python
decision_tree = DecisionTreeClassifier(criterion='entropy', random_state=0)

decision_tree.fit(x_credit_treinamento,y_credit_treinamento)
```

```python
previsoes = decision_tree.predict(x_credit_teste)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_credit_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(decision_tree)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)
```

```python
print(classification_report(y_credit_teste, previsoes))
```

```python
# print da arvore em txt e em imagem branca e preta
from sklearn import tree
import matplotlib.pyplot as plt

previsores = ['income', 'age', 'loan']

figura, eixos =  plt.subplots(nrows=1, ncols=1, figsize=(30,15))
tree.plot_tree(decision_tree, feature_names=previsores, class_names=str(decision_tree.classes_), filled=True);

figura.savefig(r'C:\Users\natha\Documents\02-Cursos\Udemy - Data Science\decision_tree\credit.png')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=40, criterion='entropy', random_state=0)

random_forest.fit(x_credit_treinamento, y_credit_treinamento)
```
